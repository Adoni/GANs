```python
from matplotlib import pyplot as plt
import seaborn as sns
import pickle
import numpy
```

```python
def plot_errors(filename, save_name=None):
    errG, errD = pickle.load(open(filename, 'rb'), encoding='latin1')
    plt.xlabel('Batch')
    plt.ylabel('Sinkhorn Loss')
    plt.plot(errG)
    if save_name:
        plt.savefig('%s_G.png'%save_name, dpi=100)
    plt.show()
    if errD[0]!=[]:
        new_err_D=[]
        for e in errD:
            new_err_D.append(-1*numpy.mean(e))
        plt.plot(new_err_D)
        plt.xlabel('Batch')
        plt.ylabel('Sinkhorn Loss')
        if save_name:
            plt.savefig('%s_D.png'%save_name, dpi=100)
        plt.show()
```

```python
plot_errors('./results/Sinkhorn_MLP/MNIST/0.10/loss.pkl')
```

```python
plot_errors('./results/Sinkhorn_MLP/MNIST/1.00/loss.pkl',save_name='./experiment/e6-mnist-curve/loss_curve_mlp_mnist')
```

```python
plot_errors('./results/Sinkhorn_DC/MNIST/1.00/loss.pkl',save_name='./experiment/e6-mnist-curve/loss_curve_dc_mnist')
```
